```python
from convexmtl_torch.data.variations_mnist import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets = FashionMNISTVariations('data', None, None)
```

```python
data, target = datasets[0]

nrows = 3
ncols = 3
fig, axs = plt.subplots(nrows, ncols, figsize=(10, 10))

for i in range(nrows):
    for j in range(ncols):
        image = data[i*ncols + j]
        label = target[i*ncols + j]
        image = np.array(image, dtype='float')
        pixels = image.reshape((28, 28))
        # plt.imshow(pixels, cmap='gray')
        axs[i, j].imshow(pixels, cmap='gray')
        axs[i, j].set_title('Title')
        # axs[i, j].tick_params(left = False, right = False , labelleft = False, labelbottom = False, bottom = False)

```
