```python
from convexmtl_torch.data.DataLoader import DataLoader
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataloader = DataLoader('data')
```

```python
df_data, df_target, _, _, task_info = dataloader.load_dataset('variations_fashionmnist', max_size=10)
```

```python
shuffle_idx = np.arange(df_data.shape[0])
np.random.shuffle(shuffle_idx)
data = df_data[shuffle_idx,:task_info]
task = df_data[shuffle_idx,task_info]
target = df_target[shuffle_idx]
data
```

```python
nrows = 3
ncols = 6
fig, axs = plt.subplots(nrows, ncols, figsize=(20, 10))

for i in range(nrows):
    for j in range(ncols):
        image = data[i*ncols + j]
        label = target[i*ncols + j]
        image = np.array(image, dtype='float')
        pixels = image.reshape((28, 28))
        # plt.imshow(pixels, cmap='gray')
        axs[i, j].imshow(pixels, cmap='gray')
        axs[i, j].set_title('L: {}, T: {}'.format(target[i*ncols + j], task[i*ncols + j]))
        # axs[i, j].tick_params(left = False, right = False , labelleft = False, labelbottom = False, bottom = False)

```
